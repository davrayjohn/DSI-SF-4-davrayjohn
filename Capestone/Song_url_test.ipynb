```python
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from scrapy.selector import Selector
from scrapy.http import HtmlResponse
import pandas as pd
from bs4 import BeautifulSoup
import json
```

```python

parsed_song = "/Machine-gun-kelly-bad-things-lyrics"
song_url = "https://genius.com%s" % parsed_song
```

```python
driver = webdriver.Chrome('../../../../DavisJohnson/Downloads/chromedriver')  # Optional argument, if not specified will search path.
driver.get(song_url)
```

```python
title, pyong, artist, album, views, contributors, featuring, produced_by, tags, lyrics = [], [], [], [], [], [], [], [], [], []

table_rows = Selector(text=driver.page_source).xpath("//div[contains(@ng-controller, 'SongCtrl')]")


for row in table_rows:
    title.append(row.xpath(".//h1[@class='song_header-primary_info-title']/text()").extract())
    pyong.append(row.xpath(".//span[@class='song_header-pyongs_count']/text()").extract())
    artist.append(row.xpath(".//a[@class='song_header-primary_info-primary_artist']/text()").extract())
    album.append(row.xpath(".//a[@ng-bind='album.name']/text()").extract())
    views.append(row.xpath(".//div[@class='metadata_unit']/text()").extract())
    contributors.append(row.xpath(".//span[@class='metadata_unit-link']/text()").extract())
    featuring.append(BeautifulSoup(str(row.xpath(".//expandable-list[@collection='song.featured_artists']").extract())).text)
    produced_by.append(BeautifulSoup(str(row.xpath(".//expandable-list[@collection='song.producer_artists']").extract())).text)
    tags.append(row.xpath(".//span[contains(@ng-class, 'metadata_unit')]/text()").extract())
    lyrics.append(BeautifulSoup(str(row.xpath(".//div[contains(@class, 'lyrics')]//lyrics").extract())).text)
    #commentary.append(row.xpath(".//standard-rich-content").extract())
    #commentary.append(row.xpath("//div[contains(@class, 'rich_text_formatting')]").extract())
    
    
```

```python
table = pd.DataFrame({
        'Title': title, 
        'Pyong': pyong,
        'Artist': artist,
        'Album': album,
        'Views': views,
        'Contributors': contributors,
        'featuring': featuring,
        'produced_by': produced_by,
        'tags': tags,
        'lyrics': lyrics})
```
